```python
import json
import pandas as pd

with open(
    "data/4_blhs_merged.json",
    "r",
    encoding="utf-8",
) as f:
    data = json.load(f)

condition = {}

for row in data:
    if row.get("condition_event"):
        condition[row["condition_event"]] = row.get(
            "condition_event_name", ""
        )

effect = {}

for row in data:
    if row.get("effect_event"):
        effect[row["effect_event"]] = row.get(
            "effect_event_name", ""
        )

pd.DataFrame(
    [
        {
            "event_id": k,
            "event_name": v,
        }
        for k, v in sorted(condition.items())
    ]
).to_csv(
    "data/condition_events.csv",
    index=False,
    encoding="utf-8-sig",
)

pd.DataFrame(
    [
        {
            "event_id": k,
            "event_name": v,
        }
        for k, v in sorted(effect.items())
    ]
).to_csv(
    "data/effect_events.csv",
    index=False,
    encoding="utf-8-sig",
)
```

```python
bridge = sorted(set(condition.keys()) & set(effect.keys()))

pd.DataFrame({
    "bridge_event": bridge
}).to_csv(
    "data/bridge_events.csv",
    index=False,
    encoding="utf-8-sig",
)
```
